# Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame({"x": data[:,0], "y": data[:,1], "class": np.random.randint(0,2,data.shape[0])})
df[:6]
```

```python
df.describe()
```

```python
df.hist()
```

```python
df[["x","y"]][:20].plot(kind='bar')
```

```python
s = df[df['class'] == 1].plot(kind='scatter', x='x', y='y', c='red', label = 'spam', alpha = 0.7)
df[df['class'] == 0].plot(kind='scatter', x='x', y='y', c='green', label = 'ham', alpha=0.7, ax = s)
```
